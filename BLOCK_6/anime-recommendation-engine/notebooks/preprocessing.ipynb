# Profiles

```python
#files 
data = pd.read_csv("https://anime-recommendation-engine.s3.eu-west-3.amazonaws.com/data/profiles.csv")
data.head()
```

```python
# Basic stats
print(f"Number of rows : {data.shape[0]}")
print(f"Number of columns : {data.shape[1]}")

print("Basics statistics: ")
display(data.describe(include='all'))

print("Percentage of missing values: ")
print(data.isna() .sum() / data.shape[0])
```

### EDA

```python
data['age'] = data["birthday"].apply(lambda x : "".join(re.findall("[0-9]{4}", x)) if pd.notnull(x) else "")
data['age'] = data['age'].replace("", "0")
data['age'] = data['age'].apply(lambda x : 2020-int(x) if 2020-int(x)<2020 else "")
data
```

### profile

```python
data['profile'].value_counts()
```

```python
data[data['profile']=="Voivodian"].head()
```

```python
df = data.drop_duplicates('profile')
```

### stats after cleaning

```python
# Basic stats
print(f"Number of rows : {df.shape[0]}")
print(f"Number of columns : {df.shape[1]}")

print("Basics statistics: ")
display(df.describe(include='all'))

print("Percentage of missing values: ")
print(df.isna() .sum() / df.shape[0])
```

```python
df.to_csv('../data/profiles_clean.csv', index=False)
```
