```python
#
import pandas as pd

import os

# 
import plotly.express as px
import plotly.graph_objects as go

#
from scipy.stats import f_oneway

# SimpleImputer : class which allow us to fill missing values (null, NaN) with specific strategy (median, mean, mode, etc.)
from sklearn.impute import SimpleImputer

pd.options.display.max_columns = None
pd.options.display.max_rows = None
```

### Loading

```python
data_path = os.getcwd() + "/data/"

dataset = pd.read_csv(data_path + "Speed+Dating+Data.csv", encoding='ISO-8859-1')
display(dataset.head())
```

```python
print(dataset.isnull().sum())
print(dataset.dtypes)
```

```python
dataset.describe()
```

```python
# Since we're going to use statisticals tests, we're using an imputer to put a mean to all missing values of type float64
# (even if some of them doesn't really make sense with a mean, we are not using them on those analysis)
imputer = SimpleImputer(strategy="mean")
dataset = dataset.copy()

columns_impute = [item for i, item in enumerate(dataset.columns) if dataset[item].dtype in ["float64"]]

dataset.loc[:,columns_impute] = imputer.fit_transform(dataset.loc[:,columns_impute])
```

```python
print(dataset.isna().sum())
```

```python
df_main_data = dataset.loc[:,:"shar1_1"]
display(df_main_data.shape)
display(df_main_data.head())
```

### Useful functions

```python
def anova_oneway(df_1, df_2):
    stat, p = f_oneway(df_1, df_2)

    if p < 0.05:
        print(f'[{df_1.name}|{df_2.name}] Reject null hypothesis: at least one group mean is different (p={p})')
    else:
        print(f'[{df_1.name}|{df_2.name}] Fail to reject null hypothesis: all group means are the same (p={p})')
```
